```python
data = pd.DataFrame({'name' : [], 'column_1' : [], 'column_2' : [],
                     'column_3' : [], 'column_4' : [], 'column_5' : [],
                     'column_6' : [], 'column_7' : [], 'column_8' : [],
                     'column_9' : [], 'column_10' : [], 'class' : []})
i = 1
while i != 210:
    name = "data/1/" + str(i) + ".png"
    image = plt.imread(name)
    gray = rgb2gray(image)
    summ = gray.sum(axis = 0)
    result_vector = []
    j = 0
    column_summ = 0
    while j != summ.shape[0]-1:
        column_summ += summ[j]
        j+=1
        if j % int(summ.shape[0]/10 + 1) == 0:
            result_vector.append(column_summ)
            column_summ = 0
    result_vector.append(column_summ)
    data = data.append({'name' : name, 'column_1' : result_vector[0], 'column_2' : result_vector[1],
                         'column_3' : result_vector[2], 'column_4' : result_vector[3], 'column_5' : result_vector[4],
                          'column_6' : result_vector[5], 'column_7' : result_vector[6], 'column_8' : result_vector[7],
                            'column_9' : result_vector[8], 'column_10' : result_vector[9], 'class' : 0},
                              ignore_index=True)
    i += 1
```
