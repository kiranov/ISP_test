```python
def data_creation(size):
    data = pd.DataFrame({})
    i = 1
    while i != 210:
        name = "data/1/" + str(i) + ".png"
        image = plt.imread(name)
        gray = rgb2gray(image)
        summ = gray.sum(axis = 0)
        result_vector = []
        j = 0
        column_summ = 0
        while j != summ.shape[0]-1:
            column_summ += summ[j]
            if len(result_vector) < size-1:
                if (j+1) % (int(summ.shape[0]/size)) == 0:
                    result_vector.append(column_summ)
                    column_summ = 0 
            j += 1
        result_vector.append(column_summ)
        line = {'name' : name}
        l = 1
        while l <= size:
            string = 'column_'
            string += str(l)
            line[string] = result_vector[l-1]
            l += 1
        line['class'] = 0
        data = data.append(line, ignore_index=True)
        i += 1
    k = 1
    while k != 212:
        name = "data/2/" + str(k) + ".png"
        image = plt.imread(name)
        gray = rgb2gray(image)
        summ = gray.sum(axis = 0)
        result_vector = []
        j = 0
        column_summ = 0
        while j != summ.shape[0]-1:
            column_summ += summ[j]
            if len(result_vector) < size-1:
                if (j+1) % (int(summ.shape[0]/size)) == 0:
                    result_vector.append(column_summ)
                    column_summ = 0 
            j += 1
        result_vector.append(column_summ)
        line = {'name' : name}
        l = 1
        while l <= size:
            string = 'column_'
            string += str(l)
            line[string] = result_vector[l-1]
            l += 1
        line['class'] = 1
        data = data.append(line, ignore_index=True)
        k += 1
    data = shuffle(data)
    return data
```

```python
def data_creation_lines(size1, size2):
    data = pd.DataFrame({})
    i = 1
    while i != 210:
        name = "data/1/" + str(i) + ".png"
        column_line = {'name' : name}
        image = plt.imread(name)
        gray = rgb2gray(image)
        line = 1
        step = int(gray.shape[0] / size2)
        nil = 0
        while line <= size2:
            if(line == size2):
                pattern = gray[nil:gray.shape[0]]
            else:
                pattern = gray[nil:step*line]
            summ = pattern.sum(axis = 0)
            result_vector = []
            j = 0
            column_summ = 0
            while j != summ.shape[0]-1:
                column_summ += summ[j]
                if len(result_vector) < size1-1:
                    if (j+1) % (int(summ.shape[0]/size1)) == 0:
                        result_vector.append(column_summ)
                        column_summ = 0 
                j += 1
            result_vector.append(column_summ)
            l1 = 1
            while l1 <= size1:
                string = 'column_' + str(line)
                string += str(l1)
                column_line[string] = result_vector[l1-1]
                l1 += 1
            column_line['class'] = 0
            nil = step * line
            line += 1
        data = data.append(column_line, ignore_index=True)
        i += 1
    k = 1
    while k != 212:
        name = "data/2/" + str(k) + ".png"
        column_line = {'name' : name}
        image = plt.imread(name)
        gray = rgb2gray(image)
        line = 1
        step = int(gray.shape[0] / size2)
        nil = 0
        while line <= size2:
            if(line == size2):
                pattern = gray[nil:gray.shape[0]]
            else:
                pattern = gray[nil:step*line]
            summ = pattern.sum(axis = 0)
            result_vector = []
            j = 0
            column_summ = 0
            while j != summ.shape[0]-1:
                column_summ += summ[j]
                if len(result_vector) < size1-1:
                    if (j+1) % (int(summ.shape[0]/size1)) == 0:
                        result_vector.append(column_summ)
                        column_summ = 0 
                j += 1
            result_vector.append(column_summ)
            l1 = 1
            while l1 <= size1:
                string = 'column_' + str(line)
                string += str(l1)
                column_line[string] = result_vector[l1-1]
                l1 += 1
            column_line['class'] = 1
            nil = step * line
            line += 1
        data = data.append(column_line, ignore_index=True)
        k += 1
    data = shuffle(data)
    return data
```
